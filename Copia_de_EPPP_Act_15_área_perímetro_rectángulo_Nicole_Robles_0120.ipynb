📝 Redacción Extendida del Problema
* IntroducciónEn el estudio de la geometría euclidiana, el rectángulo es una de las figuras fundamentales.
* Se define como un paralelogramo cuyos cuatro ángulos internos son rectos ($90^\circ$).
* La capacidad de calcular sus dimensiones no solo es un ejercicio matemático, sino una necesidad técnica en campos como el desarrollo de interfaces gráficas (GUI), donde las ventanas y botones son rectángulos, o en la ingeniería civil para el cálculo de superficies de terrenos.
* Descripción del EscenarioSe requiere desarrollar una aplicación de consola robusta utilizando el lenguaje de programación Python.
* La aplicación debe servir como una calculadora geométrica especializada. El problema principal radica en la interacción con el usuario y el procesamiento de datos numéricos.
* El sistema debe abordar los siguientes puntos:Modularidad: El código debe estar organizado en funciones independientes para facilitar su mantenimiento.
* Interfaz de Usuario (CLI): Un menú persistente que permita realizar múltiples cálculos sin necesidad de reiniciar el programa.Tratamiento de Datos: Manejo de variables de punto flotante (float) para permitir mediciones precisas (con decimales).📐
* Fundamentos Matemáticos y VisualizaciónPara comprender qué está ejecutando el código, debemos desglosar las propiedades del rectángulo:El Área ($A$): Representa la medida de la superficie encerrada dentro de los límites de la figura.
* Se calcula multiplicando sus dos dimensiones ortogonales.$$A = b \cdot h$$El Perímetro ($P$): Representa la longitud total del contorno de la figura.
* Dado que un rectángulo tiene dos bases iguales y dos alturas iguales, la fórmula se simplifica a:$$P = 2b + 2h = 2(b + h)$$

```python
# programa elaborado por Nicole Robles
import time

def calcular_area_rectangulo(base, altura):
    """💎 Calcula la superficie total del rectángulo."""
    return base * altura

def calcular_perimetro_rectangulo(base, altura):
    """📏 Calcula la longitud del contorno del rectángulo."""
    return 2 * (base + altura)

def mostrar_menu():
    """✨ Despliega una interfaz visual elegante en la consola."""
    print("\n" + "═"*40)
    print("  🌟 CALCULADORA GEOMÉTRICA PRO 🌟")
    print(" " + "═"*38)
    print("  1️⃣  Calcular Área Rectángulo")
    print("  2️⃣  Calcular Perímetro Rectángulo")
    print("  3️⃣  Finalizar Aplicación")
    print("═"*40)

    opcion = input("✨ Seleccione una opción (1-3): ")
    return opcion

def ejecutar_programa():
    """🚀 Controla el flujo lógico y la experiencia del usuario."""
    activo = True

    while activo:
        opcion = mostrar_menu()

        if opcion == '1':
            print("\n📐 [MODO ÁREA]")
            try:
                b = float(input("🔹 Ingrese la base: "))
                h = float(input("🔹 Ingrese la altura: "))
                area = calcular_area_rectangulo(b, h)
                print(f"\n✅ RESULTADO: El área es {area} u²")
            except ValueError:
                print("\n❌ ERROR: Por favor, ingrese solo números.")
            time.sleep(1.5)

        elif opcion == '2':
            print("\n📏 [MODO PERÍMETRO]")
            try:
                b = float(input("🔹 Ingrese la base: "))
                h = float(input("🔹 Ingrese la altura: "))
                perimetro = calcular_perimetro_rectangulo(b, h)
                print(f"\n✅ RESULTADO: El perímetro es {perimetro} u")
            except ValueError:
                print("\n❌ ERROR: Por favor, ingrese solo números.")
            time.sleep(1.5)

        elif opcion == '3':
            print("\n✨ Cerrando sesión... ¡Gracias por su preferencia! ✨")
            activo = False

        else:
            print("\n⚠️  OPCIÓN INVÁLIDA: Intente con 1, 2 o 3.")
            time.sleep(1)

# --- INICIO DEL PROGRAMA ---
if __name__ == "__main__":
    ejecutar_programa()
```
